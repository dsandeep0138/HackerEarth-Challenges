```python
import cv2
import pytesseract
from pathlib import Path
from pytesseract import Output
from textblob import TextBlob
import pandas as pd
```

```python
img_dir = 'Dataset'
```

```python
def getSentiment(text):
    sent = TextBlob(text).sentiment.polarity

    if sent > 0:
        return "Positive"
    elif sent == 0:
        return "Random"
    else:
        return "Negative"
```

```python
#flist = sorted([p.name for p in Path(img_dir).iterdir() if p.is_file()])

df = pd.read_csv('Sample_Submission.csv')
```

```python
sentiments = []

for path in df['Filename']:
    filepath = img_dir + "/" + path
    image = cv2.imread(filepath)
    
    text = pytesseract.image_to_string(image, config = "-l eng --oem 1 --psm 11")
    print("file: {0}, sentiment: {1}".format(path, TextBlob(text).sentiment.polarity))
    sentiments.append(getSentiment(text))
```

```python
results = pd.DataFrame({"Filename" : df['Filename'], "Category" : sentiments})
results.to_csv("results.csv", index = False)
```
